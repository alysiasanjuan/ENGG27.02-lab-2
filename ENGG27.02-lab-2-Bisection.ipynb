```julia
function f(x)
    -2*x^6-1.5*x^4+10*x+2
end;
```

```julia
f(x)
plot(f, -1, 1.5)
```

```julia
x = 1.3

f(x)
```

```julia
function bisect(xl,xu,es,imax)
    iter = 0
    fl = f(x)
    xr = xu
    ea = es + 1
    
    while ea > es && iter < imax
        xr_old = xr
        xr = (xl + xu)/2
        fr = f(x)
        iter += 1
        
        if xr != 0
            ea = abs((xr - xr_old)/xr)*100
        end
        
        test = fl*fr
        
        if test < 0
            xu = xr
        elseif test > 0
            xl = xr
            fl = fr
        else
            ea = 0
        end
    end
    
    return xr,iter,ea
end
```

```julia
xl = 0
xu = 1
es = 5
imax = 10000

bisect(xl,xu,es,imax)
```
